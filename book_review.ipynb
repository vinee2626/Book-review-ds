## 2. In which year the maximum number of books were in demand?  

```python
demand = pd.DataFrame(df.groupby(['year_of_publication']).count()['isbn'].reset_index())
demand
```

```python
#pd.DataFrame(df['year_of_publication'].value_counts()).plot(kind='bar' ,figsize=(20,8))
#plt.xticks(rotation=90)
#plt.show()
plt.figure(figsize=(20,6))
plt.scatter(demand['year_of_publication'], demand['isbn'], s=25, color= '#00CED1', edgecolors='black')
plt.xlabel('BOOKS PUBLISHED OVER THE YEARS ', size= 15)
plt.ylabel('Number of books published', size =15)
plt.show()
```

### 3. How many year old books have the best ratings? 

```python
df['rating'] = pd.to_numeric(df['rating'])
df['age'] = df['age'].astype('int64')  
df.info()
```

```python
rating = pd.DataFrame(df.groupby(['age']).max().sort_values(by= 'rating',ascending=False)['rating'].reset_index())
rating
```

```python
plt.bar(rating['age'], rating['rating'], color= '#022d36', width=0.5)
plt.xlabel('AGE OF THE BOOKS', size=10)
plt.ylabel('Rating of books', size=10)
plt.show()
```

```python
pd.DataFrame(df.loc[df['rating'] == 10,['age', 'rating']])
```

### 4.Books in what language are most preffered? And in which country? 

```python
lang = pd.DataFrame(df.groupby(['Language', 'country']).count()['isbn'].sort_values(ascending=False).reset_index())
lang
```

```python
plt.figure(figsize=(10,6))
plt.scatter(lang['isbn'], lang['Language'], facecolor='green', alpha=0.85, edgecolors='black')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()
```
